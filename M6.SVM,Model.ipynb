SVM implementation in Python
Predict if cancer is Benign or malignant. Using historical data about patients diagnosed with cancer enables doctors to differentiate malignant cases and benign ones are given independent attributes.



#Steps

1,Load the breast cancer dataset from sklearn.datasets

2,Separate input features and target variables.

3,Build and train the SVM classifiers using RBF kernel.

4,Plot the scatter plot of the input features.

5,Plot the decision boundary.

```python
'''Support Vector Machine (SVM) is a supervised machine learning algorithm
capable of performing classification, regression and even outlier detection'''

# Load the important packages

from sklearn.datasets import load_breast_cancer
import matplotlib.pyplot as plt
from sklearn.inspection import DecisionBoundaryDisplay
from sklearn.svm import SVC
import pandas as pd
import numpy as np

# Load the datasets
cancer = load_breast_cancer()
X = cancer.data[:, :2]
y = cancer.target

# print(X)

#Build the model
svm = SVC(kernel="rbf", gamma=0.5, C=1.0)

# Trained the model
svm.fit(X, y)

# Plot Decision Boundary
# A line or curve that divides the data into two or more categories based on their features.
DecisionBoundaryDisplay.from_estimator(
		svm,
		X,
		response_method="predict",
		cmap=plt.cm.Spectral,           # you can access the Spectral colormap using plt.cm.Spectral  
		alpha=0.8,						# allows you to regulate the transparency of a graph plot
		xlabel=cancer.feature_names[0],
		ylabel=cancer.feature_names[1],
	)

''' plot first Two features in a 2d Plot.  Using : means that we take all    
elements in the correspond array dimension'''
plt.scatter(X[:, 0], X[:, 1],     
			c="green",		  	  # Color
			s=50,                 # maker_size
            edgecolors="black")
plt.show()

# print(cancer)
```
